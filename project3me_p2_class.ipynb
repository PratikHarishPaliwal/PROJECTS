```python
CON=pd.DataFrame(con)
CAT=pd.DataFrame(cat)
print(CON)
print(CAT)
```

```python
from statsmodels.formula.api import ols
model = ols("Age ~ Attrition",A).fit()
from statsmodels.stats.anova import anova_lm
anova_results = anova_lm(model)
pval = list(anova_results[["PR(>F)"]].values)[0][0]
print("Pvalue: ",round(pval,4))
if(pval < 0.05):
    print("Dependent")
else:
    print("Independent")
```

```python
CON_TR=CON.T
CON_TR.columns=con
print(CON_TR)

CAT_TR=CAT.T
CAT_TR.columns=cat
print(CAT_TR)
```

```python
for i in CON_TR:
    print(i)
print("*******************************************************")
print("=======================================================")
for i in CAT_TR:
    print(i)
```

```python
imp_cols_ANOVA=[]
un_imp_cols_ANOVA=[]
for i in CON_TR:
    from statsmodels.formula.api import ols
    model = ols("A[i] ~ Attrition",A).fit()
    from statsmodels.stats.anova import anova_lm
    anova_results = anova_lm(model)
    pval = list(anova_results[["PR(>F)"]].values)[0][0]
    print("Pvalue: ",round(pval,4))
    if(pval < 0.05):
        print(i,"Dependent")
        print("********************")
        imp_cols_ANOVA.append(i)
    else:
        print(i,"Independent")
        print("====================")
        un_imp_cols_ANOVA.append(i)
```
