```python
from bs4 import BeautifulSoup
from requests import get
```

```python
data = get(url)


soup = BeautifulSoup(data.content)
```

```python
data.content#rawdata
```

```python
soup#ordered data using BeautifulSoup
```

```python
name = []
for i in soup.find_all("h1"):
    name.append(i.span.text)
```

```python
name
```

```python
brand = []
brand.append(name[0][0:6])
```

```python
brand 
```

```python
price = []
for i in soup.find_all("div",class_=['_30jeq3 _16Jk6d']):
    price.append(i.text)
```

```python
price
```

```python
Reviews = []
Q = []
for i in soup.find_all("span",class_=['_2_R_DZ']):
    Q.append(i.text)
```

```python
Reviews.append(Q[0][16:19])
```

```python
Reviews
```

```python
print("brandname:",brand,"\n","price:",price,"\n","Reviews:",Reviews)
```
